```python
import numpy as np
import matplotlib.pyplot as plt
import drag_functions_turb as fn
```

```python
impact_params_arr = np.load("Impact_Param_arrays.npy")
```

```python
r_WS_arr = np.zeros(impact_params_arr.shape[1])
r_h_arr = np.zeros(impact_params_arr.shape[1])
with open("Impact_Param_tests.txt", "r") as text_file:
    for i, line in enumerate(text_file):
        if i%2 == 1:
            r_h_arr[i//2] = line.split()[8][:-1]
            r_WS_arr[i//2] = line.split()[12]
    
plt.loglog(impact_params_arr[0], impact_params_arr[1], 'b.')
plt.loglog(impact_params_arr[0], r_h_arr, 'r--')
plt.loglog(impact_params_arr[0], r_WS_arr, 'g--')
plt.xlabel('Particle size (cm)')
plt.ylabel('Accretion radius (cm)')
```
